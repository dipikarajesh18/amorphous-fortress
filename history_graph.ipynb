```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# import the numpy history arrays
SEEDS = [171265,797894,958110,369459,778702]

best_scores = []
cur_scores = []
num_entities = []
for s in SEEDS:
    history_np = np.load(f"ALIFE_EXP/_history/history_s[{s}].npy",allow_pickle=True)
    best_scores.append(history_np[0])
    cur_scores.append(history_np[1])
    num_entities.append(history_np[2])
```

```python
# make averages of everything
best_scores = np.array(best_scores)
cur_scores = np.array(cur_scores)
num_entities = np.array(num_entities)

best_scores_avg = np.mean(best_scores,axis=0)
cur_scores_avg = np.mean(cur_scores,axis=0)
num_entities_avg = np.mean(num_entities,axis=0)

max_best_scores = np.max(best_scores,axis=0)
max_cur_scores = np.max(cur_scores,axis=0)
max_num_entities = np.max(num_entities,axis=0)

min_best_scores = np.min(best_scores,axis=0)
min_cur_scores = np.min(cur_scores,axis=0)
min_num_entities = np.min(num_entities,axis=0)
```

```python
# plot the graph with the stuffs (i'm drunk coding help)

plt_colors = ['blue','green']

plt.figure(figsize=(10,5))


plt.plot(range(0,len(best_scores_avg),10),cur_scores_avg[::10],label="Current Score", linewidth=3, color=plt_colors[0])
plt.fill_between(range(0,len(cur_scores_avg),10),min_cur_scores[::10],max_cur_scores[::10],color=plt_colors[0],alpha=0.3)

plt.plot(range(0,len(cur_scores_avg),10),best_scores_avg[::10],label="Best Score", linewidth=3, color=plt_colors[1])
plt.fill_between(range(0,len(best_scores_avg),10),min_best_scores[::10],max_best_scores[::10],color=plt_colors[1],alpha=0.3)
plt.legend(("Current Score (avg)", "Current Score (range)", "Best Score (avg)", "Best Score (range)"), loc='upper left')
plt.xlabel("Generation #")
plt.ylabel("Fitness Score")
# plt.title("Fitness Score over Generations")

plt.show()
```

```python
# plot the graph with the stuffs (i'm drunk coding help)

plt.figure(figsize=(10,5))

plt.plot(range(0,1000,10),num_entities_avg[::10],label="Num Entites", linewidth=3, color='red')
plt.fill_between(range(0,len(num_entities_avg),10),min_num_entities[::10],max_num_entities[::10],color='red',alpha=0.4)
plt.legend(("Num Entities (avg)", "Num Entities (range)"), loc='upper left')
plt.show()
```
